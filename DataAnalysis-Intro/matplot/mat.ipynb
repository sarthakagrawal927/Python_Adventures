```python
heart = pd.read_csv("Z:\Coding\PythonMagic\DataAnalysis\matplot\heart-disease.csv")
heart
```

```python
heart.head()
```

```python
heart.plot.hist(figsize = (10,30) ,subplots  = True);
```

```python
over50 = heart[heart["age"] > 50]
over50
```

```python
#pyplot
over50.plot(kind="scatter", x='age', y= 'chol', c= 'target');
```

```python
#object oriented method
fig, ax = plt.subplots(figsize = (10,6))
over50.plot(kind='scatter',x='age',y='chol',c='target',ax = ax);
ax.set_xlim([45,80]);
```

```python
fig ,ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over50["age"], y = over50["chol"], c = over50["target"]);

# add title and axis
ax.set(title="Heart didease and cholestrol levels", xlabel="Age", ylabel= "Cholesterol");

# add legends
ax.legend(*scatter.legend_elements(), title = "Target");

# adding an horizontal line
ax.axhline(over50["chol"].mean(), linestyle='--');
```

```python
fig, (ax0,ax1) = plt.subplots(nrows=2,ncols=1, figsize=(10,10), sharex=True)

scatter = ax0.scatter(x=over50["age"], y=over50["chol"], c= over50["target"]);
ax0.set(title="Cholestrol", ylabel="cholesterol");
ax0.axhline(over50["chol"].mean(), linestyle = '--');
ax0.legend(*scatter.legend_elements(),title = "target");

scatter2 = ax1.scatter(x=over50["age"], y = over50['thalach'], c = over50['target']);
ax1.set(title="THalach", xlabel="Age", ylabel="thalach");
ax1.axhline(over50["thalach"].mean(), linestyle = '--');
ax1.legend(*scatter.legend_elements(),title = "target");

#adding title to the figure
fig.suptitle("Hear disease analysis", fontweight = "bold");
```
